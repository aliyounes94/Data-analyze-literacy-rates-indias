```python
##Python package that provides fast, flexible, and expressive data structures
import pandas as pd
```

```python
## importing csv file for processing
df = pd.read_csv('literacy-rates-indias_cleaned.csv')
print(df.head())  ## show file first data
```

```python

import matplotlib.pyplot as plt
# Group by Gender and Age Group to calculate the average literacy rate
age_gender_avg = df.groupby(['Gender', 'Age Group'])['Literacy Rate (%)'].mean().reset_index()

# Pivot the data for easier visualization
age_gender_pivot = age_gender_avg.pivot(index='Age Group', columns='Gender', values='Literacy Rate (%)')

# Plot the results
age_gender_pivot.plot(kind='bar', figsize=(10, 6), color=['blue', 'pink'])
plt.title('Relationship Between Age and Literacy Rate by Gender')
plt.xlabel('Age Group')
plt.ylabel('Literacy Rate (%)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
