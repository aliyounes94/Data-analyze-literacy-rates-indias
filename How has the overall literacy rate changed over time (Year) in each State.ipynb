```python
##Python package that provides fast, flexible, and expressive data structures
import pandas as pd
```

```python
## importing csv file for processing
df = pd.read_csv('literacy-rates-indias_cleaned.csv')
print(df.head())  ## show file first data
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Group by State and Year to calculate the average literacy rate
state_year_avg = df.groupby(['State 2', 'Year'])['Literacy Rate (%)'].mean().reset_index()

# Pivot the data for easier visualization
state_year_pivot = state_year_avg.pivot(index='Year', columns='State 2', values='Literacy Rate (%)')

# Plot the trends for each state
plt.figure(figsize=(14, 8))
for state in state_year_pivot.columns:
    plt.plot(state_year_pivot.index, state_year_pivot[state], label=state)

plt.title('Change in Literacy Rate Over Time by State')
plt.xlabel('Year')
plt.ylabel('Literacy Rate (%)')
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.tight_layout()
plt.show()
```
